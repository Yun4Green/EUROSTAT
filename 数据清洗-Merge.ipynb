```python
import pandas as pd
```

# Tire Value

```python
value = pd.read_csv('./Tire/raw_data/tire_product_Value_PCR_TBR_SP.csv')
value
```

```python
value.columns
```

```python
value_product = value.groupby(['country', 'date', 'PRODUCT', 'partner', 'HS Code', 'Category', 'General_Category'])['value'].sum().reset_index()
```

```python
value_product
```

# Tire Weight 

```python
weight = pd.read_csv('./Tire/raw_data/tire_product_Weight_PCR_TBR_SP.csv')
weight
```

```python
weight_product = weight.groupby(['country', 'date', 'PRODUCT', 'partner', 'HS Code', 'Category', 'General_Category'])['weight'].sum().reset_index()
```

```python
weight_product
```

# merge

```python
eurostat = value_product.merge(weight_product,on = ['country', 'date', 'PRODUCT', 'partner', 'HS Code', 'Category', 'General_Category'] ,how='inner')
```

```python
eurostat['weight'] *=100
```

```python
eurostat['Type'] = 'Tyre'
```

```python
eurostat
```

```python
eurostat.columns
```

```python
eurostat_final = eurostat[['date', 'HS Code', 'PRODUCT', 'Category', 'General_Category', 'country', 'partner', 'value', 'weight', 'Type']]
```

```python
category_list = ['PCR', 'TBR', 'Specialty']
eurostat_product =  eurostat_final[eurostat_final['General_Category'].isin(category_list)]
```

```python
# eurostat_product = eurostat_product[(eurostat_product['date']<"2025-10-01")]
eurostat_product
```

# Vehicle Value

```python
vehicle_value = pd.read_csv('./Vehicle/Data/vehicle_product_value.csv')
vehicle_value
```

```python
vehicle_value.columns
```

```python
vehicle_value_product = vehicle_value.groupby(['country', 'date', 'PRODUCT', 'partner', 'HS Code','General_Category'])['value'].sum().reset_index()
vehicle_value_product
```

# Vehicle Weight

```python
vehicle_weight = pd.read_csv('./Vehicle/Data/vehicle_product_weight.csv')
vehicle_weight
```

```python
vehicle_weight_product = vehicle_weight.groupby(['country', 'date', 'PRODUCT', 'partner', 'HS Code','General_Category'])['weight'].sum().reset_index()
vehicle_weight_product
```

# Merge

```python
eurostat_vehicle = vehicle_value_product.merge(vehicle_weight_product,on = ['country', 'date', 'PRODUCT', 'partner', 'HS Code', 'General_Category'] ,how='inner')
```

```python
eurostat_vehicle['weight'] *=100
```

```python
eurostat_vehicle['Type'] = 'Vehicle'
```

```python
eurostat_vehicle
```
